**Let's create a new feature.**

```python
def age_category(age):
    """
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    """
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data["Age_category"] = age_categories
```

**Another way is to do it with `apply`.**

```python
data["Age_category"] = data["Age"].apply(age_category)
```

**4. Is that true that the mean age of survived people is higher than that of passengers who eventually died?**
- Yes
- No


```python
mean_age_survived = data[data['Survived'] == 1]['Age'].mean()
mean_age_died = data[data['Survived'] == 0]['Age'].mean()

if mean_age_survived > mean_age_died:
    print("Yes")
else:
    print("No")
```

**5. Is that true that passengers younger than 30 y.o. survived more frequently than those older than 60 y.o.? What are shares of survived people among young and old people?**
- 22.7% among young and 40.6% among old
- 40.6% among young and 22.7% among old
- 35.3% among young and 27.4% among old
- 27.4% among young and  35.3% among old

```python
young = data[data['Age'] < 30]
old = data[data['Age'] > 60]

young_survived_rate = young['Survived'].mean() * 100
old_survived_rate = old['Survived'].mean() * 100

if young_survived_rate > old_survived_rate:
    print("True")
else:
    print("False")

print(f"Survival rate among young passengers (<30): {young_survived_rate:.1f}%")
print(f"Survival rate among old passengers (>60): {old_survived_rate:.1f}%")
```

**6. Is that true that women survived more frequently than men? What are shares of survived people among men and women?**
- 30.2% among men and 46.2% among women
- 35.7% among men and 74.2% among women
- 21.1% among men and 46.2% among women
- 18.9% among men and 74.2% among women

```python
men = data[data['Sex'] == 'male']
men_survived_rate = men['Survived'].mean() * 100

women = data[data['Sex'] == 'female']
women_survived_rate = women['Survived'].mean() * 100

if women_survived_rate > men_survived_rate:
    print("True")
else:
    print("False")

print(f"Survival rate among men: {men_survived_rate:.1f}%")
print(f"Survival rate among women: {women_survived_rate:.1f}%")
```

**8. How is average age for men/women dependent on `Pclass`? Choose all correct statements:**
- On average, men of 1 class are older than 40
- On average, women of 1 class are older than 40
- Men of all classes are on average older than women of the same class
- On average, passengers ofthe first class are older than those of the 2nd class who are older than passengers of the 3rd class

```python
mean_ages = data.groupby(['Sex', 'Pclass'])['Age'].mean()
mean_ages
```

```python
men_class_1_age = mean_ages['male'][1] > 40
print("Men of 1st class are older than 40:", men_class_1_age)

women_class_1_age = mean_ages['female'][1] > 40
print("Women of 1st class are older than 40:", women_class_1_age)

men_older_than_women = all(mean_ages['male'][pclass] > mean_ages['female'][pclass] for pclass in [1, 2, 3])
print("Men are older than women in the same class:", men_older_than_women)

class_age_order = mean_ages.loc[('male', 1)] > mean_ages.loc[('male', 2)] > mean_ages.loc[('male', 3)] and \
                  mean_ages.loc[('female', 1)] > mean_ages.loc[('female', 2)] > mean_ages.loc[('female', 3)]
print("Passengers of 1st class are older than 2nd, who are older than 3rd:", class_age_order)
```
